# K Nearest Neighbor(KNN)

Notice: You should find the best k to build the model with the best accuracy.\
**warning:** You should not use the **loan_test.csv** for finding the best k, however, you can split your train_loan.csv into train and test to find the best **k**.


```python
def transform_df(df):
    
    df =df.copy()
    df.drop(['Unnamed: 0.1','Unnamed: 0'],axis=1,inplace=True)
    df['due_date'] = pd.to_datetime(df['due_date'])
    df['effective_date'] = pd.to_datetime(df['effective_date'])
    df['dayofweek'] = df['effective_date'].dt.dayofweek
    df['weekend'] = df['dayofweek'].apply(lambda x: 1 if (x>3)  else 0)
    df['from_effective_to_due_date'] = (df.due_date - df.effective_date).dt.days
    df.drop(['effective_date','due_date'],axis=1,inplace=True)
    df['Female']= df['Gender'].replace(to_replace=['male','female'], value=[0,1])
    df.drop('Gender',axis=1,inplace=True)
    df['PAIDOFF'] = df['loan_status'].replace(to_replace=['COLLECTION','PAIDOFF'], value=[0,1])
    df.drop('loan_status',axis=True,inplace=True)
    dummies =  ['terms','education']
    df = pd.get_dummies(df,columns= dummies)
    sc_features =   MinMaxScaler()
    numerical_features = ['Principal','age','from_effective_to_due_date','dayofweek']
    df[numerical_features]  = sc_features.fit_transform(df[numerical_features])
    return df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import jaccard_score,accuracy_score,f1_score,confusion_matrix, classification_report
transformed_df= transform_df(df)
features = list(filter(lambda x: x!='PAIDOFF', transformed_df.columns))
target='PAIDOFF'
X =transformed_df[features].values
y =transformed_df[target].values
X_train, X_test, y_train, y_test = train_test_split(
   X, y, test_size=0.25,random_state=13
)
```

```python
from sklearn.neighbors import KNeighborsClassifier
ks_list = np.arange(2,11)
Ks_score = np.zeros((len(ks_list)))

for n,k in enumerate(ks_list):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    Ks_score[n] =  accuracy_score(y_test, yhat)
```

```python
plt.plot(ks_list,Ks_score,'g')
plt.ylabel('accuracy')
plt.xlabel('Number of Neighbors (K)')
plt.tight_layout()
plt.show()
print( "The best accuracy_score was with", Ks_score.max(), "with k=",ks_list[Ks_score.argmax()]) 
```

```python
KN = KNeighborsClassifier(n_neighbors = k).fit(X,y)
```

# Decision Tree


```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
from sklearn.tree import DecisionTreeClassifier
import sklearn.tree as tree
tree_depths = np.arange(2,30)
tree_results = np.zeros((len(tree_depths)))
for n,d in enumerate(tree_depths):
    loanTree = DecisionTreeClassifier(criterion="entropy", max_depth = d)
    loanTree.fit(X_train,y_train)
    loanPredTree = loanTree.predict(X_test)
    tree_results[n] = accuracy_score(y_test,loanPredTree)
```

```python
 print("DecisionTrees's Accuracy: ", tree_results.max(), "with dept:", tree_depths[tree_results.argmax()])
```

```python
loanTree = DecisionTreeClassifier(criterion="entropy", max_depth = 2).fit(X,y)
plt.figure(figsize=(10,10))
tree.plot_tree(loanTree)
plt.show()
```

```python
cnf_matrix = confusion_matrix(y_test, loanPredTree , labels=[0,1])
np.set_printoptions(precision=2)
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=['Collection','PAIDOFF'],normalize= False,  title='Confusion matrix')
```
